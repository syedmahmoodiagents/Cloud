```python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
```

```python
import os, getpass
```

```python
os.environ["AWS_ACCESS_KEY_ID"] = getpass.getpass("Access key")
os.environ["AWS_SECRET_ACCESS_KEY"] = getpass.getpass("Secret access key")
os.environ["AWS_DEFAULT_REGION"] = "us-east-1"
```

```python
from langchain_aws import ChatBedrock
```

```python
llm = ChatBedrock(
    model_id="amazon.nova-micro-v1:0",   # or amazon.titan, etc.
    region_name="us-east-1"
)
```

```python
prompt = ChatPromptTemplate.from_messages([
    ("system", "Use the retrieved context to answer the user."),
    ("human", "{question}\n\nContext:\n{context}")
])
```

```python
chain = prompt | llm
```

```python
response = chain.invoke({"question": "What is Amazon Bedrock?", "context": "Act as if you have expert in AWS "})
```
